```python
data = pd.read_csv(files)
filename = files.split('.csv')[0]
col = ['Admin2','Province_State', 'Country_Region', 'Confirmed',
               'Deaths', 'Recovered']
data = data[col]
data.columns = ['County','Province_State', 'Country_Region', 'Confirmed_'+filename,
               'Deaths_'+filename, 'Recovered_'+filename]
data = data.fillna(' ')
data['Country_State_County'] = data['Country_Region']+'_'+data['Province_State']+'_'+data['County']
data = data.drop(['Country_Region','Province_State','County'],axis=1)
data = data.groupby(['Country_State_County']).sum()
print(data)
i=0
for file in files[:-1]:
    filename = file.split('.csv')[0]
    df1 = pd.read_csv(file)
    if i<60:
        df1['Admin2'] = ' '
        col = ['Admin2','Province/State', 'Country/Region', 'Confirmed',
               'Deaths', 'Recovered']
        df1 = df1.replace({'Country/Region':{'Mainland China':'China'}})
    else:
        col = ['Admin2','Province_State', 'Country_Region', 'Confirmed',
               'Deaths', 'Recovered']
    df1 = df1[col]
    df1.columns = ['County','Province_State', 'Country_Region', 'Confirmed_'+filename,
               'Deaths_'+filename, 'Recovered_'+filename]
    df1 = df1.fillna(' ')
    df1['Country_State_County'] = df1['Country_Region']+'_'+df1['Province_State']+'_'+df1['County']
    df1 = df1.drop(['Country_Region','Province_State','County'],axis=1)
    df1 = df1.groupby(['Country_State_County']).sum()
    #print(df1)
    data = pd.merge(data,df1,how='left', on=['Country_State_County'])
    #print(i,len(df1))
    i+=1
    #break
data
```
